```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
df=pd.read_csv("C:\\Users\\pratiksha\\Documents\\Excelar\\data science assinements\\PCA\\wine.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Considering only numerical data 
df.data = df.iloc[:,1:]
df.data.head()
```

```python
# Converting into numpy array
df1 = df.data.values
df1
```

```python
# Normalizing the numerical data 
wine_normal = scale(df1)
```

```python
wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type')
```

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
dendrogram=sch.dendrogram(sch.linkage(wine_normal,'complete'))
```

```python
# Create Clusters
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
x=pd.DataFrame(hclusters.fit_predict(wine_normal),columns=['clustersid'])
x['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
wine2=df.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

### Kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

### Build cluster Algoritham withk=3

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(wine_normal)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
wine3=df.copy()
wine3['clusters3id']=clusters3.labels_
wine3
```
